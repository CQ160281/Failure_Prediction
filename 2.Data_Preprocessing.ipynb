## 5. Feature 

### 5.1 Add Features

```python
# Add features làm cho thời gian cảm biến mượt hơn bằng cách thêm rolling mean (Đường trung bình động)
# và standard deviation(độ lệch chuẩn)
def add_features(df_in, rolling_win_size):
    
    ''' Add rolling average and rolling standard deviation cho cảm biến, 
    sử dụng cố định rolling window size.
            
           df_in (dataframe):  inppput của dataframe sẽ được sử lí (training or test). 
           rolling_win_size (int): window size, số chu kỳ áp dụng cho rolling function.
           
           dataframe: contains chứa inpput dataframe thêm vào với rolling mean and std cho cảm biến.
    
    '''
    
    sensor_cols = ['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11',
                's12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
    
    sensor_av_cols = [nm.replace('s', 'av') for nm in sensor_cols]
    sensor_sd_cols = [nm.replace('s', 'sd') for nm in sensor_cols]
    
    df_out = pd.DataFrame()
    
    ws = rolling_win_size
    
    # Tính rolling stats cho id động cơ
    
    for m_id in pd.unique(df_in.id):
    
        # get subset (tập con) cho cảm biến của động cơ
        df_engine = df_in[df_in['id'] == m_id]
        df_sub = df_engine[sensor_cols]

    
        # get rolling mean cho subset
        av = df_sub.rolling(ws, min_periods=1).mean()
        av.columns = sensor_av_cols
    
        # get the rolling standard deviation cho tập con
        sd = df_sub.rolling(ws, min_periods=1).std().fillna(0)
        sd.columns = sensor_sd_cols
    
        # merge 2 cột tập con mới vào dataframe 
        new_ftrs = pd.concat([df_engine,av,sd], axis=1)
    
        # add thêm hàng cho features mới vào output dataframe
        df_out = pd.concat([df_out,new_ftrs])
        
    return df_out
```

### 5.2 Add the regression and classification labels training data

```python
def prepare_train_data (df_in, period):
    
    ''' Thêm regression and classification labels cho training data.
            
            Regression label: ttf (time-to-failure) = last cycle - cycle
            Binary classification label: label_bnc = if ttf <= chu kỳ parameter then 1 else 0 (values = 0,1)
        
        Args:    
            df_in (dataframe): The input training data
            period (int): Sổ cycles để phân loại TTF dùng classification labels để phân loại
               
        
        Returns: 
            dataframe: Input dataframe with regression and classification labels đã được thêm.
    '''
    
    # Tạo regression label 
    df_max_cycle = pd.DataFrame(df_in.groupby('id')['cycle'].max())
    df_max_cycle.reset_index(level=0, inplace=True)
    df_max_cycle.columns = ['id', 'last_cycle']
    
    #Thêm  mới cột time-to-failure ttf 
    df_in = pd.merge(df_in, df_max_cycle, on='id')
    df_in['ttf'] = df_in['last_cycle'] - df_in['cycle']
    df_in.drop(['last_cycle'], axis=1, inplace=True)
    
    #Tạo binary classification label
    df_in['label_bnc'] = df_in['ttf'].apply(lambda x: 1 if x <= period else 0)
    
    #Tạo multi-class classification label
    df_in['label_mcc'] = df_in['ttf'].apply(lambda x: 2 if x <= period/2 else 1 if x <= period else 0)

    return df_in
```

### 5.3 Add the regression and classification labels test data

```python
def prepare_test_data(df_test_in, df_truth_in, period):
    
    '''Add regression and classification labels cho test data.
    
             Regression label: ttf (time-to-failure) = extract cycle cuối cho động cơ và merge nó với truth data
             Binary classification label: label_bnc = if ttf <= chu kỳ parameter then 1 else 0 (values = 0,1)
             
        Args:
          df_in (dataframe): The input training data
          period (int): Sổ cycles để phân loại TTF dùng classification labels để phân loại
          
        Returns: 
            dataframe: Input dataframe with regression and classification labels đã được thêm.
          
    '''
    
    df_tst_last_cycle = pd.DataFrame(df_test_in.groupby('id')['cycle'].max())
    
    df_tst_last_cycle.reset_index(level=0, inplace=True)
    df_tst_last_cycle.columns = ['id', 'last_cycle']
    
    df_test_in = pd.merge(df_test_in, df_tst_last_cycle, on='id')


    df_test_in = df_test_in[df_test_in['cycle'] == df_test_in['last_cycle']]

    df_test_in.drop(['last_cycle'], axis=1, inplace=True)
    
    df_test_in.reset_index(drop=True, inplace=True)
    
    df_test_in = pd.concat([df_test_in, df_truth], axis=1)
    
    #Tạo binary classification label
    df_test_in['label_bnc'] = df_test_in['ttf'].apply(lambda x: 1 if x <= period else 0)
    
     #Tạo multi-class classification label
    df_test_in['label_mcc'] = df_test_in['ttf'].apply(lambda x: 2 if x <= period/2 else 1 if x <= period else 0)


    return df_test_in
```
